```python
filtered_data[numerical_cols] = filtered_data[numerical_cols].replace([np.inf, -np.inf], np.nan)
filtered_data[numerical_cols] = filtered_data[numerical_cols].fillna(filtered_data[numerical_cols].median())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
filtered_data['Timestamp'] = pd.to_datetime(filtered_data['Timestamp'])


filtered_data['Hour'] = filtered_data['Timestamp'].dt.hour
filtered_data['DayOfWeek'] = filtered_data['Timestamp'].dt.day_of_week
filtered_data['Minute'] = filtered_data['Timestamp'].dt.minute

# Drop the original Timestamp column (optional)
filtered_data = filtered_data.drop(columns=['Timestamp'])
filtered_data = filtered_data.drop(columns=['Source IP', 'Destination IP'])
def categorize_port(port):
    if 0 <= port <= 1023:
        return 'Well-known Port'
    elif 1024 <= port <= 49151:
        return 'Registered Port'
    elif 49152 <= port <= 65535:
        return 'Dynamic Port'
    else:
        return 'Unknown Port'

filtered_data['Source Port Category'] = filtered_data['Source Port'].apply(categorize_port)
filtered_data['Destination Port Category'] = filtered_data['Destination Port'].apply(categorize_port)
filtered_data = filtered_data.drop(columns=['Source Port', 'Destination Port'])

# Step 5: Encode Protocol
filtered_data = pd.get_dummies(filtered_data, columns=['Protocol', 'Source Port Category', 'Destination Port Category'], drop_first=True)
scaler=MinMaxScaler()
numcols=[ 'Flow Duration','Total Fwd Packets', 'Total Backward Packets', 'Flow Bytes/s',
       'Flow Packets/s', 'Packet Length Mean', 'Packet Length Std',
       'SYN Flag Count', 'ACK Flag Count', 'FIN Flag Count',
       'RST Flag Count', 'Flow IAT Mean', 'Flow IAT Std']
filtered_data[numcols]=scaler.fit_transform(filtered_data[numcols])
label=LabelEncoder()
filtered_data['Label']=label.fit_transform(filtered_data['Label'])
import pickle
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(label, f)

```

```python
from sklearn.utils import resample
normal_data=filtered_data[filtered_data['Label']==1]
ddos_data=filtered_data[filtered_data['Label']==0]
normal_data_sampled=resample(normal_data,replace=True,n_samples=100000,random_state=42)
ddos_data_sampled=resample(ddos_data,replace=False,n_samples=100000,random_state=42)
balanced_dataset=pd.concat([normal_data_sampled,ddos_data_sampled])
```
